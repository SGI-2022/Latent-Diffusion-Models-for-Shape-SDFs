```python
import os 
main_dir = './models/'
all_logs = []

for sub_dir in os.scandir(main_dir):
    if sub_dir.is_dir():
        for file in os.listdir(main_dir + sub_dir.name):
            if file.endswith('.log'):
                
                with open(os.path.join(main_dir, sub_dir.name, file)) as f:
                    f = f.readlines()

                for line in f:
                    if "load_all_shapes" in line:
                        load_all_shapes = line.split(' = ')[1].split(',')[0]
                    if "n_shapes" in line:
                        n_shapes = line.split(' = ')[2].split(',')[0].strip()
                        if load_all_shapes == 'True':
                            n_shapes = 6778
                    if "dropout" in line:
                        dropout = line.split('=')[3].split(',')[0].strip()
                    if "weight_norm" in line:
                        weight_norm = line.split('=')[5].split(',')[0].strip()
                    if "use_tanh" in line:
                        use_tanh = line.split('=')[6].strip()
                    if "n_points_to_load" in line:
                        n_points_to_load = line.split(' = ')[2].split(',')[0]
                    if "batch_size" in line:
                        batch_size = line.split(' = ')[3].split(',')[0]
                    if "n_epochs" in line:
                        n_epochs = line.split(' = ')[5].split(',')[0]
                    if "loss" in line and 'Validation' not in line:
                        epoch_trained = line.split('-')[2].split('loss')[0].strip()

                try:
                    all_logs.append([sub_dir.name, n_shapes, dropout, weight_norm, use_tanh, n_points_to_load, batch_size, epoch_trained])

                except Exception:
                    pass

```

```python
import numpy as np
import pandas as pd

all_logs = np.array(all_logs)
df = pd.DataFrame(all_logs, columns = ['datetime','n_shapes', 'dropout', 'weight norm', 'tanh', 'n_subsmaples', 'batch size', 'epochs trained'])
df = df.sort_values(by=['datetime'])
df = df.set_index('datetime')
df["remarks"] = ""
```
