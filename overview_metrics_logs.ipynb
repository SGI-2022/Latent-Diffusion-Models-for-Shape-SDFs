```python
import os
import pandas as pd
import numpy as np

```

```python
main_dir = './diffusion logs & models/'

# metrics_logs_paths = ['./generated_set_normalized/metrics.log'] 
# print(metrics_logs_paths[0])
metrics_logs_paths= []
for sub_dir in os.scandir(main_dir):
    if sub_dir.is_dir():
        if (sub_dir.name)[:2] != '11': # we are interested in recent logs only
            continue
        for sub_sub_dir in os.scandir(main_dir + sub_dir.name):
            if 'generated shapes' in sub_sub_dir.name:
                metrics_log_path = os.path.join(main_dir, sub_dir.name, sub_sub_dir.name, 'metrics.log')
                if os.path.exists(metrics_log_path):
                    print(metrics_log_path)
                    metrics_logs_paths.append(metrics_log_path)

```

```python
results = []
# for log in ['./diffusion logs & models/11062022_220611/generated shapes-11062022_220611-2500/metrics.log']:
for log in metrics_logs_paths:
    with open(log) as f:
        f = f.readlines()
    
    filename_dt = log.split('/')[2].strip()
        
    info = '' # reset all 
    MMD = ''
    COV = ''
    NNA = ''
    decoder_type = ''
    dataset = ''
    n_timestep = ''
    diffusion_epoch = ''
    diffusion_dt = ''
    print(log)
    for idx, line in enumerate(f):
        if (idx == 1019) and (log == './generated_set_normalized/metrics.log'): # use the first metric result for this log
            break
        if "Using " in line:
            if line.split('Using ')[1].strip().split(',')[0].strip() != filename_dt: 
                break # some logs have metrics results saved for subsequent runs

            info = line.split('Using ')[1].strip()
            if info.split(',')[2].strip() == './data/03001627_sdfs/':
                dataset = '50k chairs'
            elif info.split(',')[2].strip() == '/mnt/disks/data2/latent_diffusion/03001627_sdfs_500k':
                dataset = '500k chairs'
                
            diffusion_dt = info.split(',')[0].strip()
            diffusion_epoch = info.split(',')[1].strip()
            
            # print(f[idx+1])
            # decoder_type = f[idx+1].split('-')[2].split(',')[0].strip()
            # n_timestep = f[idx+1].split('-')[2].split(',')[3].strip()

        if "MMD: " in line:
            MMD = line.split('MMD: ')[1].split('[')[1].split(']')[0].strip()
        if "COV: " in line:
            COV = line.split('COV: ')[1].strip()
        if "1NNA: " in line:
            NNA = line.split('1NNA: ')[1].strip()
            # print(diffusion_dt, MMD, COV, NNA)
    results.append([diffusion_dt, info, MMD, COV, NNA, dataset, diffusion_epoch])
    # print(results)
        
results = np.array(results)
df = pd.DataFrame(results, columns = ['diffusion_dt', 'info', 'MMD', 'COV', 'NNA', 'dataset', 'diffusion epoch'])
df = df.sort_values(by=['diffusion_dt'])
df = df.set_index('diffusion_dt')

df["remarks"] = ""
df.at["11022022_030925", "remarks"] = "final, DeepSDF"
df.at["11042022_142710", "remarks"] = "1000 timesteps"
df.at["11082022_135018", "remarks"] = "final, our autodecoder"

# df.at[0, "remarks"] =

df

#TODO: separate into 100 shapes and 500 shapes results
```
